```python
# this function help to repalce the M, K and 'Varies with device' into suitable form

def size_process(item):
    if str(item)[-1] == "M":
        res = float(str(item).replace("M", " "))
        res = res * 1000
        return str(res)
    if str(item)[-1] == "k":
        res = str(item).replace("k", "")
        return res
    else:
        return str(np.nan)
```

```python
# repalcing the M, K and 'Varies with device' into a suitable form of 'Size' column

df['Size'] = df['Size'].apply(size_process)
```

```python
# changing the data type into  float values

df['Size'] = df['Size'].astype(float)
```
